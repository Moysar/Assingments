##**Guess the Number Game**
**Description:**
The "Guess the Number Game" is a simple Python project designed to introduce
students to basic programming concepts such as user input, random number
generation, conditionals, and loops. The game provides an interactive experience where
the computer generates a random number, and the player has to guess that number within a specified range.

Key Features:
1. **Random Number Generation:**
 - The program generates a random number between a specified range (e.g., 1 to 100).
2. **User Input:**
 - Players input their guesses for the randomly generated number.
 - The program provides feedback on whether the guess is too high, too low, or correct.
3. **Conditional Statements:**
 - The program uses if-else statements to check if the player's guess is correct and provides appropriate feedback.

4. **Loop Structure:**
 - The game is structured using a loop to allow multiple guesses until the correct
number is guessed.
5. **Counting Attempts:**
 - The program keeps track of the number of attempts it takes the player to guess the
correct number.
**Usage:**
- Players run the program and are prompted to input their guesses.
- The program provides feedback after each guess, guiding the player to the correct answer.
- The game continues until the player correctly guesses the number.

#**Bonus:**
- Adding difficulty levels with different ranges.
- Implementing a scoring system based on the number of attempts.
- Incorporating a limited number of guesses.

```python
%%timeit -n 5 -r 1

import random

def guess_the_number():
    print("Welcome to the Guess the Number Game!")
    print("Choose the difficulty level: ")
    print("1. Easy (1-10) - 5 Guesses")
    print("2. Medium (1-100) - 10 Guesses")
    print("3. Hard (1-1000) - 15 Guesses")
    level = input("Enter your choice: ")

    if level == "1":
        lower_bound, upper_bound, max_attempts = 1, 10, 5
    elif level == "2":
        lower_bound, upper_bound, max_attempts = 1, 100, 10
    elif level == "3":
        lower_bound, upper_bound, max_attempts = 1, 1000, 15
    else:
        print("Invalid choice. Defaulting to Easy.")
        lower_bound, upper_bound, max_attempts = 1, 10, 5

    secret_number = random.randint(lower_bound, upper_bound)
    attempts = 0
    guessed_correctly = False

    print(f"\nI'm thinking of a number between {lower_bound} and {upper_bound}.")
    print(f"You have {max_attempts} attempts to guess it.")

    while attempts < max_attempts and not guessed_correctly:
        try:
            remaining = max_attempts - attempts
            user_input = input(f"\n[Attempts remaining: {remaining}] Enter your guess: ")
            guess = int(user_input)
            attempts += 1

            if guess < secret_number:
                print("Too low! Try again.")
            elif guess > secret_number:
                print("Too high! Try again.")
            else:
                guessed_correctly = True
                print(f"\nCongratulations! You've guessed the correct number: {secret_number}")
                print(f"It took you {attempts} attempts.")
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

    if not guessed_correctly:
        print(f"\nGame Over! You've run out of attempts.")
        print(f"The secret number was: {secret_number}")

if __name__ == "__main__":
    guess_the_number()
```
